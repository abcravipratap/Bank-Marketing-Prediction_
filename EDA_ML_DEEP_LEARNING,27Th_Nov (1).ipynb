```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data['job']=le.fit_transform(data['job'])
data['marital']=le.fit_transform(data['marital'])
data['education']=le.fit_transform(data['education'])
data['default']=le.fit_transform(data['default'])
data['housing']=le.fit_transform(data['housing'])
data['loan']=le.fit_transform(data['loan'])
data['contact']=le.fit_transform(data['contact'])
data['month']=le.fit_transform(data['month'])
data['poutcome']=le.fit_transform(data['poutcome'])
data['deposit']=le.fit_transform(data['deposit'])
```

```python
data.head(5)
```

```python
data.columns
```

```python
x=data[['age', 'job', 'marital', 'education', 'default', 'balance', 'housing',
       'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays',
       'previous', 'poutcome']]
y=data[['deposit']]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2,random_state=40)

```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
from sklearn.svm import SVC
model= SVC()
model.fit(xtrain,ytrain)
```

```python
ypred= model.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred)
```

```python
from sklearn.linear_model import LogisticRegression
model1= LogisticRegression()
model1.fit(xtrain,ytrain)
ypred1= model1.predict(xtest)
ypred1
accuracy_score(ytest,ypred1)
```

```python
from sklearn.ensemble import RandomForestClassifier
model2= RandomForestClassifier(n_estimators=200)
model2.fit(xtrain,ytrain)
ypred2= model2.predict(xtest)
ypred2
accuracy_score(ytest,ypred2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model= Sequential()
```

```python
model.add(Dense(32,activation='relu',input_dim=16))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(xtrain,ytrain,epochs=50,batch_size=10,validation_data=(xtest,ytest))
```

```python
model.summary()
```

```python
y_ped=model.predict(xtest)
y_ped
y_pred=[]
for i in y_ped:
  if i>0.5:
    y_pred.append(1)
  else:
      y_pred.append(0)


```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,y_pred)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
model5= Sequential()
model5.add(Dense(32,activation='relu',input_dim=16))
model5.add(Dense(32,activation='relu'))
model5.add(Dense(1,activation='sigmoid'))
model5.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# without early stopping
history=model5.fit(xtrain,ytrain,epochs=500,batch_size=50,validation_data=(xtest,ytest))
y_ped5=model5.predict(xtest)
y_ped5
y_pred5=[]
for i in y_ped5:
  if i>0.5:
    y_pred5.append(1)
  else:
      y_pred5.append(0)
from sklearn.metrics import accuracy_score
accuracy_score(ytest,y_pred5)

```

```python
# with early stopping
```

```python
model6= Sequential()
model6.add(Dense(32,activation='relu',input_dim=16))
model6.add(Dense(32,activation='relu'))
model6.add(Dense(1,activation='sigmoid'))
model6.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python

from tensorflow.keras.callbacks import EarlyStopping
es=EarlyStopping(monitor='val_loss',patience=50)
history=model6.fit(xtrain,ytrain,epochs=500,batch_size=100,validation_data=(xtest,ytest),callbacks=[es])
y_ped6=model6.predict(xtest)
y_ped6
y_pred6=[]
for i in y_ped6:
  if i>0.5:
    y_pred6.append(1)
  else:
      y_pred6.append(0)
from sklearn.metrics import accuracy_score
accuracy_score(ytest,y_pred6)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
